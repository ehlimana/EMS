```python
import os
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file=pd.read_excel("when2heat.xlsx",parse_dates=True,header=[0,1,2,3])
```

```python
d=[]
e=[]
from datetime import datetime
for value1,value2 in zip(file.iloc[:,0],file.iloc[:,1]):
    value1=value1.replace('Z', '')
    value2=value2.replace('Z', '')
    value1,value2=datetime.fromisoformat(value1),datetime.fromisoformat(value2)
    value1.strftime('%Y-%m-%d %H:%M:%S')
    value2.strftime('%Y-%m-%d %H:%M:%S')
    d.append(value1),e.append(value2)
```

```python
file.iloc[:,0]=d
file.iloc[:,1]=e
```

```python
file.iloc[:,0]=pd.to_datetime(file.iloc[:,0],utc=True)
file.iloc[:,1]=pd.to_datetime(file.iloc[:,1],utc=True)
```

```python
file=file.dropna(subset=[(     'HR',  'heat_demand',         'space',          'MW')],axis=0)
```

```python
heat_demand=[('HR',  'heat_demand',         'total',          'MW'),(     'HR',  'heat_demand',         'water',          'MW')]
file.plot(x=('country',     'variable',     'attribute',        'unit'),y=heat_demand,kind="line",figsize=(20,10))
plt.show()
```

```python
file=file.drop(file.index[[0,1]])
```

```python
%matplotlib inline
from statsmodels.tsa.seasonal import seasonal_decompose
fig, ax = plt.subplots(ncols=4, nrows=2, sharex=True, figsize=(15,15))

for i,column in enumerate(heat_demand):
    decomp = seasonal_decompose(x=file[column], 
                                model='additive',period=24)
    ax[i,0].set_title('Decomposition of {}'.format(column), fontsize=7)
    decomp.observed.plot(ax=ax[i,0], legend=False, color='dodgerblue')
    ax[i,0].set_ylabel('Observed', fontsize=7)

    decomp.trend.plot(ax=ax[i,1], legend=False, color='dodgerblue')
    ax[i,1].set_ylabel('Trend', fontsize=7)
    decomp.seasonal.plot(ax=ax[i,2], legend=False, color='dodgerblue')
    ax[i,2].set_ylabel('Seasonal', fontsize=7)
    decomp.resid.plot(ax=ax[i,3], legend=False, color='dodgerblue')
    ax[i,3].set_ylabel('Residual', fontsize=7)
plt.show()
```

```python
file["Year"]=file.iloc[:,0].dt.year
file["Month"]=file.iloc[:,0].dt.month
file["Dayofyear"]=file.iloc[:,0].dt.dayofyear
file["Dayofmonth"]=file.iloc[:,0].dt.day
file["Hour"]=file.iloc[:,0].dt.hour
```

```python
year=[2008,2009,2010,2011,2012,2013]
len(file.loc[file["Year"]==2008])
```

```python
from numpy import polyfit
X = file.iloc[:,-1]
y = file.iloc[:,15]
degree =7
coef = polyfit(X, y, degree)
print('Coefficients: %s' % coef)
```

```python
# create curve
curve = list()
for i in range(len(X)):
 
    value = coef[-1]
    for d in range(degree):
    
        value+= X.values[i]**(degree-d) * coef[d]
    curve.append(value)
# plot curve over original data

plt.plot(file['HR',  'heat_demand',         'total',          'MW'].values[:24])
plt.plot(curve[:24], color='red', linewidth=3)
plt.show()
```

```python
years=[2008,2009,2010,2011,2012,2013]
file["Hourofyear"]=0
for year in years:
    hours=len(file.loc[file["Year"]==year])
    hours_range=np.arange(0,hours,1)
    file.loc[file["Year"]==year,"Hourofyear"]=hours_range
```
